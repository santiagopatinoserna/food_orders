```python
# Importing libraries
import pandas as pd

# import for visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("../data/onlinefoods.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of seaborn for better aesthetics
sns.set_style("whitegrid")

# Create subplots for the distribution by gender and age group
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

# Distribution by Gender
sns.countplot(x='Gender', data=data, palette='coolwarm', ax=ax[0])
ax[0].set_title('Distribution of Online Food Orders by Gender')
ax[0].set_xlabel('Gender')
ax[0].set_ylabel('Number of Orders')

# Distribution by Age Group
# Categorize age into groups for better visualization
bins = [0, 18, 30, 40, 50, 60, 70, 100]
labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '61-70', '71+']
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

sns.countplot(x='Age Group', data=data, palette='viridis', ax=ax[1])
ax[1].set_title('Distribution of Online Food Orders by Age Group')
ax[1].set_xlabel('Age Group')
ax[1].set_ylabel('Number of Orders')

# Display the plots
plt.tight_layout()
plt.show()
```

```python
# Monthly Income vs. Online Food Ordering
plt.figure(figsize=(10, 6))
sns.countplot(y='Monthly Income', data=data, palette='Set2', order=data['Monthly Income'].value_counts().index)
plt.title('Online Food Orders by Monthly Income')
plt.xlabel('Number of Orders')
plt.ylabel('Monthly Income')
plt.show()

# Occupation and Online Food Ordering Frequency
plt.figure(figsize=(10, 6))
sns.countplot(y='Occupation', data=data, palette='Set3', order=data['Occupation'].value_counts().index)
plt.title('Online Food Orders by Occupation')
plt.xlabel('Number of Orders')
plt.ylabel('Occupation')
plt.show()
```
